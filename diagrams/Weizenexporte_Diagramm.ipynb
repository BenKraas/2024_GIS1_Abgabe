```python
import matplotlib.pyplot as plt
import numpy as np

# Daten
categories = ['Russland', 'USA', 'Kanada', 'Frankreich', 'Ukraine', 'Australien', 'Argentinien', 'Deutschland']
values1 = [37.3, 26.1, 26.1, 19.8, 18.1, 10.4, 10.2, 9.3]  # Menge in Mio. Tonnen
values2 = [18.8, 13.2, 13.2, 10, 9.1, 5.2, 5.1, 4.7]  # Marktanteil in %

# Abbildungsgre anpassen
fig, ax1 = plt.subplots(figsize=(10, 6))

# Balkenbreite und Positionen
x = np.arange(len(categories))
width = 0.4

# Erste Reihe von Balken (linke y-Achse)
bars1 = ax1.bar(x - width / 2, values1, width, color='skyblue', edgecolor='black', label='Weizenexporte in Mio. Tonnen')
ax1.set_ylabel('Weizenexporte in Mio. Tonnen', color='Black')
ax1.tick_params(axis='y', labelcolor='black')

# Hintergrund hinzufgen
ax1.grid(axis='y', linestyle='--', linewidth=0.7, alpha=0.7)  # Horizontale Linien

# Zweite Reihe von Balken (rechte y-Achse)
ax2 = ax1.twinx()  # Zweite Achse teilen
bars2 = ax2.bar(x + width / 2, values2, width, color='orange', edgecolor='black', label='Marktanteil in %')
ax2.set_ylabel('Marktanteil in %', color='black')
ax2.tick_params(axis='y', labelcolor='black')

# Anpassen der Achsen
max_value = max(max(values1), max(values2))  # Maximalwert beider Datenstze
buffer = 0.1 * max_value  # 10% Puffer
ax1.set_ylim(0, max_value + buffer)  # Linke Achse anpassen
ax2.set_ylim(0, max_value + buffer)  # Rechte Achse anpassen

# x-Achse und Kategorien
ax1.set_xticks(x)
ax1.set_xticklabels(categories, rotation=0, ha='center')

# Titel und Legende
ax1.set_title('')
fig.legend(loc='upper left', bbox_to_anchor=(0.6, 0.9))

# Layout optimieren und anzeigen
fig.tight_layout()
plt.show()
```
